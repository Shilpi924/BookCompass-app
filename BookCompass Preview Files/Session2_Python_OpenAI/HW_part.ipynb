Goal:
 Practice real-world usage of lists and dictionaries in Python.
Task:
 Write a Python script that:
Creates a list of dictionaries, where each dictionary represents a movie:
collection = [
    {"title": "Inception", "genre": "Sci-Fi", "year": 2010},
    {"title": "The Matrix", "genre": "Action", "year": 1999},
    {"title": "Coco", "genre": "Animation", "year": 2017},
]
Prints each item in this format:
 "Inception" is a Sci-Fi movie released in 2010.

Adds a search feature:
 Ask the user for a genre and print all movies that match the entered genre.

```python
collection = [
    {"title": "Inception", "genre": "Sci-Fi", "year": 2010},
    {"title": "The Matrix", "genre": "Action", "year": 1999},
    {"title": "Coco", "genre": "Animation", "year": 2017},
    {"title": "Zootopia", "genre": "Animation", "year": 2017}
]
for movie in collection :
    print(f"{movie.get('title')} is a {movie.get('genre')} movie released in {movie.get('year')}")
    
genre=input("Enter a genre to search for movies :")

for movie in collection :
    if movie.get('genre').lower()==genre.lower():
        print(f"{movie.get('title')}")
        
```
