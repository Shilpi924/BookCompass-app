```python
import os
import textwrap
from dotenv import load_dotenv
from openai import OpenAI
system_prompts = {'strict':'Yo are strict tutor',
                          'teacherlike' :'you are a helpful assistant.',
                          'friendly':'you are a friendly tutor',
                          'creative': 'you are a creative tutor'
                          }
load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
#my_api_key[:5], my_api_key[-10:]
client = OpenAI(api_key=my_api_key)
print (my_api_key)
```

```python
def get_response_from_model(prompt, context):
    
    response = client.chat.completions.create(

        model="gpt-4.1-nano",
        messages=[
            {
                "role": "system", 
                "content": "{system_prompts.get(style,'teacherlike')}"
            },
            {
                "role": "user", 
                "content": f"Context information:\n{context}\n\nQuestion: {prompt}"
            }
        ]
    )
    return textwrap.fill(response.choices[0].message.content)

```

```python

context = ""
prompt = input("Ask something: ")

# Corrected function call
response_text = get_response_from_model(prompt, context)

print(f"\nOpenAI says: {response_text}")
```
