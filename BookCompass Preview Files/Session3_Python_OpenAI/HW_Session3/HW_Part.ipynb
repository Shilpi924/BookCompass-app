1. OpenAI Integration + Listing Assistant v0
Build a simple Python application that takes a messy commercial real estate property description and uses OpenAI to generate:
Structured JSON output
A short marketing blurb
Clarifying questions when important information is missing
Requirements:
Use Python 3.10+
Use OpenAI API
Read input using input()
Do not hallucinate missing data
Use null or empty strings when information is unavailable
Test the application with:
A clean input
A messy input
A contradictory input
Also experiment with 3 different system prompts:
Strict extraction engine
Friendly assistant
Expert broker tone
Write a short note describing:
What changed between outputs
Which prompt worked best
Why the outputs differed

Userprompt -

Retail space in Austin, 3000 sq ft, newly built, lease info missing


```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
SYSTEM_PROMPT = """
    You are a Listing Assistant.  Output Structured JSON 
    add A short marketing blurb
    and Clarifying questions when important information is missing
Do not hallucinate missing data
Use null or empty strings when information is unavailable
    """
STRICT_LISTING_ASSISTANT = """
You are a commercial real estate Listing Assistant.

Your job is to convert messy property descriptions into structured listing data.

You must:
- Extract ONLY factual information from the input
- Never guess or hallucinate missing details
- Use null for missing numbers
- Use "" for missing text fields
- Keep output clean and structured

You also generate:
- A short neutral property summary (not marketing-heavy)
- Clarifying questions for missing listing details

Return ONLY valid JSON.
"""

FRIENDLY_LISTING_ASSISTANT = """
You are a friendly commercial real estate listing assistant.

Your job is to help users clean up messy property descriptions into structured listings.

You must:
- Extract factual data from the text
- Be helpful and easy to understand
- Do not guess missing information
- Return ONLY valid JSON

Also include:
- A simple, clear property summary
- Helpful clarifying questions when needed
"""

BROKER_LISTING_ASSISTANT = """
You are a senior commercial real estate listing broker.

Your job is to transform raw property descriptions into polished listing-ready data.

You must:
- Extract factual listing information
- Maintain professional broker tone
- Do not invent missing details
- Return ONLY valid JSON

Also generate:
- A compelling marketing blurb suitable for listings
- Targeted clarifying questions to improve deal quality
"""




load_dotenv(override=True, dotenv_path="../../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=my_api_key)
```

Firt System Prompt

Userprompt -
Retail space in Austin, 3000 sq ft, newly built, lease info missing

```python

def get_response_from_model(prompt):

    response = client.chat.completions.create(

        model="gpt-4.1-nano",

        messages=[
            {
                "role": "system",
                "content": SYSTEM_PROMPT
            },
            {
                "role": "user",
                "content": f"Question: {prompt}"
            }
        ],
        response_format={"type": "json_object"},

        temperature=0
    )

    # Get JSON string
    response = response.choices[0].message.content
    return response

prompt = input("Ask something: ")

# Corrected function call
response_text = get_response_from_model(prompt)

print(f"\nOpenAI says: {response_text}")
```
