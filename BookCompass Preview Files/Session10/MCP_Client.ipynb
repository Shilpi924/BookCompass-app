```python
import requests
from fastmcp import FastMCP

app = FastMCP("Library_mcp_app")

API_KEY = "AIzaSyA9nO5cCv8TmQLbT0_so3vpwl2kBt_cLGg"

@app.tool()
def search_books(title: str):
    response = requests.get(
        "https://www.googleapis.com/books/v1/volumes",
        params={
            "q": title,
            "maxResults": 5,
            "key": API_KEY
        }
    )

    data = response.json()

    books = []

    for item in data.get("items", []):
        volume = item.get("volumeInfo", {})

        books.append({
            "title": volume.get("title"),
            "authors": volume.get("authors", []),
            "publisher": volume.get("publisher", "Unknown"),
            "publishedDate": volume.get("publishedDate", "Unknown"),
            "description": volume.get("description", "")[:300],
            "averageRating": volume.get("averageRating", "No rating"),
            "thumbnail": volume.get("imageLinks", {}).get("thumbnail", "")
        })

    return books


# For Jupyter testing, DO NOT run app.run()
# Just call:
# search_books("Harry Potter")
```
