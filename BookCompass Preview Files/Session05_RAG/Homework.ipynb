```python
from sentence_transformers import SentenceTransformer
model= SentenceTransformer('all-MiniLM-L6-v2')
document_corpus = [
    "a big query",
    "a small query",
    "a medium query",
    "some random text",
    "some more random text",
    "more text on medium query",
    "medium some more text",
    "medium query is important"
]
query = "a medium query"
corpus_vector=model.encode(document_corpus)
query_vector=model.encode(query)
#print(corpus_vector)
#print(query_vector)
top_k_docs= get_top_k_docs(query_vector, corpus_vector, k=3)
top_k_docs
```

```python
import numpy as np

def get_top_k_docs(query, embeddings, k):
    similarity = model.similarity(query, embeddings).detach().cpu().numpy().flatten()

    top_k_indices = np.argsort(similarity)[::-1][:k]

    print("Similarity scores:", similarity)
    print("Top k indices:", top_k_indices)
    top_k_docs=[document_corpus[i] for i in top_k_indices]
    return top_k_docs
```

4. Error Handling
Write a function that:
Reads a text file
Returns the file length
Handles the case where the file does not exist


```python
def check_file_exiszts(file_path):
    try:
        with open(file_path,'r') as f:
            return len(f.read())
    except FileNotFoundError:
            print(f"File '{file_path}' does not exist.")
            return None  
check_file_exiszts('file2.txt')
```
