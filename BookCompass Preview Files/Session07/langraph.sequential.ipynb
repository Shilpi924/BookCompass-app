```python
from langgraph.graph import StateGraph, START, END
def node_a_agent(dict):
 
    print("Agent A is running")
   

def node_b_agent(dict):

    print("Agent B is running")
  

builder = StateGraph(dict)

builder.add_node("Image_Recognition", node_a_agent)
builder.add_node("Recommendation_Agent", node_b_agent)

builder.add_edge(START, "Image_Recognition")
builder.add_edge("Image_Recognition", "Recommendation_Agent")
builder.add_edge("Recommendation_Agent", END)


graph = builder.compile()
print(graph.get_graph().draw_ascii())
Initial_state = {}
graph.invoke(Initial_state)
```
